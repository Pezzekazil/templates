In this repo we are going to explore different matplotlib visualisations.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from matplotlib import style
import seaborn
style.use('dark_background')
print(plt.style.available)
```

```python
# creating data
def create_data(size):
    x=range(0,size)
    y=np.random.normal(10,6,size)
    return x, y
def create_data2(size):
    x=np.random.normal(10,6,size)
    y=np.random.normal(10,6,size)
    return x, y
```

```python
fig = plt.figure()
# first number = number of rows in the layout
# second number = number of columns in the layout
# third number = position (counting from top left to bottom right)
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232)
ax3 = fig.add_subplot(233)
ax4 = fig.add_subplot(212)

x,y = create_data(20)
x1,y1 = create_data(20)
x2,y2 = create_data(20)
x3,y3 = create_data2(60)

ax1.plot(x,y)
ax2.bar(x1,y1)
ax3.plot(x2,y2)
ax4.scatter(x3,y3)

ax1.xaxis.set_label_text('hello')
ax2.xaxis.set_label_text('this is')
ax3.xaxis.set_label_text('cool')

ax2.axes.get_yaxis().set_visible(False)
ax3.axes.get_yaxis().set_visible(False)
# ax1.spines['left'].set_color('w')
# ax1.spines['left'].set_visible(True)
# ax1.spines['left'].set_linewidth(1)
# ax1.spines['bottom'].set_color('w')
# ax1.spines['bottom'].set_visible(True)
# ax1.spines['bottom'].set_linewidth(1)
# ax2.spines['bottom'].set_color('w')
# ax2.spines['bottom'].set_visible(True)
# ax2.spines['bottom'].set_linewidth(1)

ax1.grid(b=False)
ax2.grid(b=False)
ax3.grid(b=False)
ax4.grid(b=False)



plt.subplots_adjust(hspace=0.4)
ax2.set_title('Best chart ever')

plt.show();
```
