```python
days = [1,2,3,4,5]
sleeping = [7,8,6,11,7]
eating =[2,3,4,3,2]
working = [7,8,7,2,2]
playing = [8,5,7,8,13]

plt.plot([],[],color='m', label='Sleeping', linewidth=5)
plt.plot([],[],color='c', label='eating')
plt.plot([],[],color='b', label='working')
plt.plot([],[],color='y', label='playing')


plt.stackplot(days, sleeping, eating, working, playing, colors=['m','c','b','y'])
plt.legend();
```

```python
fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

x = np.random.normal(50,10,1000)
y = np.random.binomial(800,0.8,1000)

ax1.scatter(x,y, label = 'exes', color='w', s=5, marker='x')
for label in ax1.xaxis.get_ticklabels():
    label.set_rotation(45)
ax1.grid(True, color='g', linestyle='-')
ax1.yaxis.label.set_color('r')
ax1.set_yticks([x for x in range(600,680,10)])


plt.title("scatter plot")
plt.subplots_adjust(left=0.4,bottom= 0.09, right=0.91, top=0.95, wspace=0.2, hspace=0)
plt.xlabel('x')
plt.ylabel('y')
;
```

```python

d1 = date(2016, 4, 15)
d2 = date(2016, 9, 15)

delta = d2 - d1

x = [d1 + td(days=i) for i in range(delta.days + 1)]
y = np.random.normal(40,3,len(x))

fig = plt.figure()
ax1 = plt.subplot2grid((1,1),(0,0))

ax1.plot(x,y, color='b')
#ax1.fill_between(x,y,np.mean(y), color='b', alpha=0.3)
ax1.fill_between(x,y,np.mean(y), where=(y > y[0]), color='y', alpha=0.3)
ax1.fill_between(x,y,np.mean(y), where=(y < y[0]), color='r', alpha=0.3)
;
```

```python
def create_data(size):
     x=range(0,size)
     y=np.random.normal(10,6,size)
     return x, y
def create_data2(size):
     x=np.random.normal(10,6,size)
     y=np.random.normal(10,6,size)
     return x, y


fig = plt.figure()
ax1 = fig.add_subplot(231)
ax2 = fig.add_subplot(232, sharey=ax1)
ax3 = fig.add_subplot(233, sharey=ax1)
ax4 = fig.add_subplot(212, sharey=ax1)

x,y = create_data(20)
x1,y1 = create_data(20)
x2,y2 = create_data(20)
x3,y3 = create_data2(60)

ax1.plot(x,y)
ax2.bar(x1,y1)
ax3.plot(x2,y2)
ax4.scatter(x3,y3)

ax1.xaxis.set_label_text('hello')
ax2.xaxis.set_label_text('this is')
ax3.xaxis.set_label_text('cool')
for position in ['top', 'bottom', 'left', 'right']:
    for ax in [ax1,ax2,ax3,ax4]:
        ax.spines[position].set_visible(False)

ax2.axes.get_yaxis().set_visible(False)
ax3.axes.get_yaxis().set_visible(False)
plt.subplots_adjust(hspace=0.4)
ax2.set_title('Best chart ever', color='r')

plt.show();
```

```python

%matplotlib notebook

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def randrange(n, vmin, vmax):
    return (vmax-vmin)*np.random.rand(n) + vmin

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
n = 100
for c, m, zl, zh in [('r', 'o', -60, -25), ('b', '^', -30, -5)]:
    xs = randrange(n, 23, 50)
    ys = randrange(n, 0, 100)
    zs = randrange(n, zl, zh)
    ax.scatter(xs, ys, zs, c=c, marker=m)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```
